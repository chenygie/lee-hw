# Q1
1. 讀取words.txt中的所有英文單字，英文單字之間皆由
<space>做分隔
2. 按照單字出現的順序，給予編號（0, 1, 2 ...）。 (不要跳號)
3. 計算單字出現的次數。
4. 將得到編號和次數，由單字出現在words.txt的順序輸出至
Q1.txt，最後一行不要換行，每一行皆為：
<單字><space><編號><space><出現次數>

```python
with open("words.txt") as f:
    words = f.read().strip().split()
```

```python
result = []
for i in range(len(words)):
    if words.index(words[i]) == i:
#         print("{},{},{}".format(words[i],i,words.count(words[i])))
        result.append([words[i],i,words.count(words[i])])
with open("Q1.txt","w") as f:
    f.write(str(result))
```

```python
from PIL import Image
import matplotlib.pyplot as plt
img = Image.open("westbrook.jpg")
# plt.imshow(img)
pix = img.load()
for x in range(img.width):
    for y in range(img.height):
        r,g,b = pix[x,y]
        pix[x,y] = r//2,g//2,b//2
plt.imshow(img)
```

```python
%%timeit
from PIL import Image
img = Image.open("westbrook.jpg")
old_img = img.getdata()
new_img = list(map(lambda rgb : (rgb[0]//2,rgb[1]//2,rgb[2]//2),old_img))
img.putdata(new_img)
plt.imshow(img)
```

```python

import matplotlib.pyplot as plt
from PIL import Image
img = Image.open("westbrook.jpg")
new_img = Image.eval(img,lambda i : i//2)
plt.imshow(new_img)
plt.show()
```
